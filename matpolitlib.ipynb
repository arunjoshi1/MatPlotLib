```python

width = 0.25;
dev_x = [1,2,4,5,6]
dev_y = [7,8,9,1,9]
dev_legend = 'line for dev'
# plot one line
x_indexes = np.arange(len(dev_x))

plt.bar(x_indexes-width,dev_y,label=dev_legend,width=width)
#  plot 2nd line
line_y=[20,30,33,39,40]
line_legend = 'line for test'
plt.bar(x_indexes,line_y,label=line_legend,width=width)
#  style
plt.title('for test')
plt.xlabel('x lable')
plt.ylabel('y lable')
plt.legend()

plt.xticks(ticks=x_indexs,label=dev_x)
plt.show()   
```
